1. Tentukan library yang digunakan

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing  import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

2. Load dataset

```python
diabetes_dataset = pd.read_csv('data_Q.csv')
```

```python
diabetes_dataset.head()
```

```python
diabetes_dataset.shape
```

```python
diabetes_dataset['DM'].value_counts()
```

```python
# memisahkan data dan label
X = diabetes_dataset.drop(columns='DM', axis=1)
Y = diabetes_dataset['DM']
```

```python
print(X)
```

```python
print(Y)
```

3. Standarisasi data

```python
Scaler = StandardScaler()
```

```python
Scaler.fit(X)
```

```python
standarized_data = Scaler.transform(X)
```

```python
print(standarized_data)
```

```python
X = standarized_data
Y = diabetes_dataset['DM']
```

```python
print(X)
print(Y)
```

4. Memisahkan data training dan data testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

5. Membuat data latih menggunakan algortima SVM

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

6. Membuat model evaluasi untuk mengukur tingkat akurasi

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)


print('Akurasi data training adalah : ', training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)


print('Akurasi data testing adalah  : ', test_data_accuracy)
```

7. Membuat model prediksi

```python
input_data = (32.39, 32.39, 31.81, 31.53, 32.79, 32.7, 32.36, 32.08)

input_data_as_numpy_array = np.array(input_data)

input_data_reshape = input_data_as_numpy_array.reshape(1, -1)

std_data = Scaler.transform(input_data_reshape)
print(std_data)


prediction = classifier.predict(std_data)
print(prediction)

if(prediction[0]  == 0):
    print('Pasien tidak terkena diabetes')
else :
    print('Pasien terkena diabetes')
```
